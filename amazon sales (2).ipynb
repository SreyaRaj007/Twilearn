# Data Transformation

```python
# Changing the data type 
sales['Order Date'] = pd.to_datetime(sales['Order Date'])
sales['Ship Date'] = pd.to_datetime(sales['Ship Date'])     
```

```python
sales.info()
```

```python
sales['Region'] = sales['Region'].astype(str)
sales['Country'] = sales['Country'].astype(str)
sales['Item Type'] = sales['Item Type'].astype(str)
sales['Sales Channel'] = sales['Sales Channel'].astype(str)
sales['Order Priority'] = sales['Order Priority'].astype(str)
```

```python
# Adding column to dataframe which contain only month, year and month with year 
sales['Order Month'] = sales['Order Date'].dt.month
sales['Order Year'] = sales['Order Date'].dt.year
sales['Order Date MonthYear'] = sales['Order Date'].dt.strftime('%Y-%m')
sales = sales.drop(columns=['Order Date'])     
```

```python
# Adding column to dataframe which contain only month, year and month with year 
sales['Ship Month'] = sales['Ship Date'].dt.month
sales['Ship Year'] = sales['Ship Date'].dt.year
sales['Ship Date MonthYear'] = sales['Ship Date'].dt.strftime('%Y-%m')
sales = sales.drop(columns=['Ship Date'])     
```

```python
data=sales
```

### Checking Outliers

```python
plt.figure(figsize = (15,7))
sns.boxplot(data = sales.select_dtypes(exclude = 'object'))
plt.show()
```

```python
# Calculate Monthly Sales Trend
monthly_sales = data.groupby('Order Month')['Total Revenue'].sum()

# Calculate Yearly Sales Trend
yearly_sales = data.groupby('Order Year')['Total Revenue'].sum()

# Calculate Year-Monthly Sales Trend
yearly_monthly_sales = data.groupby(['Order Year', 'Order Month'])['Total Revenue'].sum()

# Plot the Monthly Sales Trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')

# Plot the Yearly Sales Trend
plt.figure(figsize=(10, 5))
plt.bar(yearly_sales.index, yearly_sales.values)
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')

# Plot the Year-Monthly Sales Trend
yearly_monthly_sales.unstack().plot(figsize=(12, 6), marker='o')
plt.title('Year-Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')

# Show the plots
plt.show()
```

From the graph above year monthly sales trend can be visualised more easily.And from this we can say that Year Monthly sales in more in the month July in the year 2012

```python
# Group the data for analysis
# Calculate Monthly Sales Trend
monthly_sales = data.groupby('Order Month')['Total Revenue'].sum()

# Calculate Yearly Sales Trend
yearly_sales = data.groupby('Order Year')['Total Revenue'].sum()

# Calculate the growth rate
yearly_growth_rate = yearly_sales.pct_change() * 100

# Plot the trends and growth rate
plt.figure(figsize=(15, 8))

# Monthly Sales Trend
plt.subplot(2, 2, 1)
monthly_sales.plot(title='Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')

# Yearly Sales Trend
plt.subplot(2, 2, 2)
yearly_sales.plot(title='Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')

# Yearly-Monthly Sales Trend
plt.subplot(2, 2, 3)
data.groupby('Order Date MonthYear')['Total Revenue'].sum().plot(title='Yearly-Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')

# Sales Growth Rate
plt.subplot(2, 2, 4)
yearly_growth_rate.plot(title='Sales Growth Rate')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

Sales trend goes high in 2012 and it come to lowest in 2015
February is the month with highest sales and August is the month with least sales
Sales growth rate was increasing from 2011 and reached the highest point at 2012 and it started to decrease and reached the lowest point at 2013.From 2013 onwards the peak formed due to growth rate was not much as before.
